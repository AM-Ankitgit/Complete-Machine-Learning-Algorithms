## 1) Ordinal Encoder

```python
it is use for input features only (Sequence)
```

```python
df1 = df.copy()
df1.head()
```

```python
df1["normalized-losses"].value_counts().to_dict()
```

```python
df1["normalized-losses"].replace({'?': np.nan},inplace=True)
```

```python
df1["normalized-losses"].isna().sum()
```

```python
df1["normalized-losses"].dtype
```

```python
df1["normalized-losses"] = df1["normalized-losses"].astype(float)
df1["normalized-losses"].dtype
```

```python
df1.head()
```

```python
df1["normalized-losses"] = df1["normalized-losses"].fillna(df1["normalized-losses"].mean()).astype(int)
```

```python
df1.head()
```

```python
df1["num-of-doors"].unique()
```

```python
df1.isna().sum()
```

```python
df1.replace({'?': np.nan},inplace=True)
```

```python
df1.isna().sum()
```

```python
df1["num-of-doors"].unique()
```

```python
df1["num-of-doors"].replace({np.nan:"two"},inplace=True)
```

```python
df1["num-of-doors"].unique()
```

```python
oe = OrdinalEncoder(categories=[["two","four"]])  # 0 1 2
oe.fit(df1[["num-of-doors"]])
df1["num-of-doors"] = oe.transform(df1[["num-of-doors"]])
```
