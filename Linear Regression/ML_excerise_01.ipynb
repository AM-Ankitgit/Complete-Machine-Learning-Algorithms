# Linearity

````python
Linear relationship between independent variable and dependent variable
correlation >>  It indicates how closely the relation of two variables

Coefficient of corelation  >> Rvalue >(Pearson correlation coefficient (Rvalue))
>> it is the indiaction of how strong the relationship is with variables
> It is one of the feature selection technique

R-value :  -1 to +1
    corr =  Covariance / Prod of std
    
std = np.sqrt(sum((Xi-Xmean)^2)/N)

R-value :  -1 to +1
Positive corr >>  R is positive
Negative corr > R is negative

Good Predictors >>  R >= 0.7 or R<=-0.7
Bad Predictors  >>  -0.3 <R < +0.3
````

# Gradient Descent Algorithm

````python
MSE > COST FUNCTION for linear regression algorithm

It is an iterative optimization algorithm to find minimum of cost function

1. It uses partial derivatives to find new m and c values
2. We will get best m & c values from infinite number of possibilities 
3. Step size become smaller and smaller till we reach global minima point
4. Global minima >> Best m & c values , Best MSE (least MSE)
````

# Best Fit Line!

````python
1. Lowest MSE
2. It passes through maximum number of data points
3. Best m & c
4. GD algorithm find one best fit line from infinite number of possibilities
````

# Overfitting

````python
Training dataset accuracy  97%
Testing dataset accuracy   78%


Low Bias and High Variance


Bias 
Low bias  > High training dataset accuracy
high bias > Low training dataset accuracy


Variance > difference between two datasets accuracies
1. High variance
    a) High Training dataset accuracy  and low Testing dataset accuracy 
    b) Low Training dataset accuracy  and High Testing dataset accuracy 
    
    
2. Low variance
    a) High Training dataset accuracy  and High Testing dataset accuracy 
    b) Low Training dataset accuracy   and low Testing dataset accuracy   
````

# How to avoid Overfitting

````python
1.Hyperparameter Tuning >  it is use to reduce variance
  GridSearchCV
  RandomizedSearchCV
2. Train with more data (5000 rows  >>> )
3. Reduce features ( 34 >> 5)
   PCA 
   Feature selection technique
4. Pruning ( Cutting branches of decision tree) >> Tree Based Model / algorithm

5. Regularization (L1 and L2)  >>> Only use for Linear model (Linear regression and Logistic regression)
   1 L1 >> Lasso Regression
   2.L2 >> Ridge Regression

6. Cross Validation
   > K-fold Cross Validation
    
7. Remove Outliers
````

# Linear Regressor
# implementation of gradient descent algorithms

````python
class Linear_Regressor():
    # intitating parameter 
    
    def __init__(self,learning_rate,no_of_iterations):
        self.learning_rate   = learning_rate
        self.no_of_iterations = no_of_iterations
    
    def fit(self,x,y):
        
        self.m,self.n= x.shape   # m as number of training data point i.e row
                                 # n as number feature i.e columns,#  hyper parameter>> learning_rate,no_of_parameter
        
        
        
        self.w = np.zeros(self.n)
        self.b = 0
        self.x=x
        self.y=y
        
        
        #implementation of gradient descent algorithms
        for i in range(self.no_of_iterations):
            self.update_weights()
    
    def update_weights(self):
        y_prediction = self.predict(self.x)
        # cal gradient 
        dw =-(2*(self.x.T).dot(self.y-y_prediction))/self.m
        db =-2*np.sum(self.y-y_prediction)/self.m
         
        # updating weights and bias
        self.w =self.w-self.learning_rate*dw
        self.b =self.b-self.learning_rate*db
        
        
        
        
    
    def predict(self,x):
        
        return x.dot(self.w)+self.b     ## formula  y=wx+b
    
````
